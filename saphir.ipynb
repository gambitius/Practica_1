```python
import numpy as np
import matplotlib.pyplot as plt

# Leer datos desde el archivo
data = np.loadtxt('ivr.txt')

# Separar las columnas en x y y
x = data[:, 0]
y = data[:, 2]

# Grado del polinomio de ajuste
grado_polinomio = 3  # Puedes ajustar este valor

# Ajuste polinÃ³mico usando NumPy
coefficients = np.polyfit(x, y, grado_polinomio)

# Crear un polinomio a partir de los coeficientes
polinomio = np.poly1d(coefficients)

# Calcular los valores de y para el polinomio ajustado
y_ajustado = polinomio(x)

# Calcular la derivada del polinomio
derivada = np.polyder(polinomio)

# Calcular los valores de y para la derivada
y_derivada = derivada(x)
plt.figure(figsize=(10,6))
plt.grid(True)
# Graficar los datos, el polinomio ajustado y su derivada
plt.scatter(x, y, label='Datos')
plt.plot(x, y_ajustado, color='red', label=f'Ajuste polinÃ³mico (grado {grado_polinomio})')
#plt.plot(x, y_derivada, color='green', label='Derivada del ajuste')
plt.xlabel('Vout')
plt.ylabel('Id')
plt.legend()
plt.show()
print("El valor de la derivada en cada punto es:", y_derivada)
```
